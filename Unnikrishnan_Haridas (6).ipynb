# Request Data And Fetching API Response

Defines the payload to be sent in the POST request, including the variant, list conditions, and applied filters. Then, it uses the fetch_data function to retrieve the data and prints the response for debugging.

```python
data = {
    "variant": "listings",
    "listConditions": {
        "relatedBrand": "winter_in_dublin_338099f",
        "category": "whats_on"
    },
    "pageNum": 0,
    "appliedFilters": {}
}

```

```python
# Define the fetch_data function
def fetch_data(url, headers, data):
    try:
        # Make a POST request
        response = requests.post(url, headers=headers, json=data)
        response.raise_for_status()  # Raise an error for bad status codes
        return response.json()  # Parse the JSON response
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None
```

```python
# Fetch the data
response_data = fetch_data(url, headers, data)

# Print the result for debugging
print(response_data)
```

# Data Transformation
Purpose: Apply transformations to the DataFrame.

Transformations:

* Convert startDate and endDate to datetime.

* Calculate duration_days.

* Classify admission_type as 'free' or 'paid'.

* Extract category_text and subcategory_text if available.

```python
# Transformations
df['startDate'] = pd.to_datetime(df['startDate'])
df['endDate'] = pd.to_datetime(df['endDate'])
df['duration_days'] = (df['endDate'] - df['startDate']).dt.days
df['admission_type'] = df['terms'].apply(lambda x: 'free' if 'free_admission' in x else 'paid')
df['category_text'] = df['category'].apply(lambda x: x.get('text') if isinstance(x, dict) else None)
df['subcategory_text'] = df['subcategory'].apply(lambda x: x.get('text') if isinstance(x, dict) else None)
# Display the DataFrame
df.head(4)
```

# Data Selection and Visualization
* Purpose: Select relevant columns and visualize event durations by category and admission type.
* Steps:
1. Select columns like headerText, startDate, endDate, duration_days, location, etc.
2. Create a bar plot showing event durations categorized by category_text and admission_type.



```python
# Select relevant columns for visualization
transformed_df = df[['headerText', 'startDate', 'endDate', 'duration_days', 'location', 'admission_type', 'category_text', 'subcategory_text']]

# Visualization 1: Event durations by category
plt.figure(figsize=(10, 6))
sns.barplot(data=transformed_df, x='category_text', y='duration_days', hue='admission_type', palette='viridis')
plt.title('Event Durations by Category and Admission Type', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Duration (days)', fontsize=12)
plt.legend(title='Admission Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Pie Chart Visualization

Create a pie chart to display the percentage of free and paid events.



```python
# Visualization 2: Count of events by admission type
plt.figure(figsize=(6, 6))
transformed_df['admission_type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=90)
plt.title('Distribution of Events by Admission Type', fontsize=14)
plt.ylabel('')  # Hides the 'Count' label
plt.show()
```

# Count Plot Visualization

Display the number of events by subcategory


```python
# Visualization 3: Number of events by subcategory
plt.figure(figsize=(10, 6))
sns.countplot(data=transformed_df, y='subcategory_text', order=transformed_df['subcategory_text'].value_counts().index, palette='coolwarm')
plt.title('Number of Events by Subcategory', fontsize=14)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Subcategory', fontsize=12)
plt.tight_layout()
plt.show()
```

# Unit Testing for Database Content and Integration

```python
import sqlite3
import unittest
import os

class TestDatabaseContent(unittest.TestCase):

    def setUp(self):
        # Setup database connection
        self.db_name = 'visitdublin_data.db'
        self.conn = sqlite3.connect(self.db_name)
        self.cursor = self.conn.cursor()

        # Create the table if it does not exist and insert mock data for testing
        self.cursor.execute('''
            CREATE TABLE IF NOT EXISTS visitdublin_data (
                id INTEGER PRIMARY KEY,
                name TEXT
            )
        ''')
        self.cursor.execute('''
            INSERT INTO visitdublin_data (name)
            VALUES ("visitdublin details")
        ''')
        self.conn.commit()

    def tearDown(self):
        # Cleanup: Drop the table and close the connection
        self.cursor.execute('DROP TABLE IF EXISTS visitdublin_data')
        self.conn.close()
        if os.path.exists(self.db_name):
            os.remove(self.db_name)  # Delete the database file to clean up

    def test_database_content(self):
        # Fetch data from the Visit dublin database
        self.cursor.execute('SELECT * FROM visitdublin_data')
        rows = self.cursor.fetchall()

        # Assert that the table contains data after initial insertion
        self.assertGreater(len(rows), 0, "The database should contain at least one row.")

    def test_data_insertion(self):
        # Insert new data and verify if it was inserted correctly
        self.cursor.execute('INSERT INTO visitdublin_data (name) VALUES ("More Details")')
        self.conn.commit()

        # Fetch the updated data
        self.cursor.execute('SELECT * FROM visitdublin_data')
        rows = self.cursor.fetchall()

        # Assert that data has been inserted correctly
        self.assertGreater(len(rows), 1, "The database should contain more than one row after insertion.")
        print(f"Successfully fetched data: {rows}")

    def test_full_integration(self):
        # Insert data and verify after insertion
        self.test_data_insertion()

        # Verify that the inserted data is present
        self.cursor.execute('SELECT * FROM visitdublin_data WHERE name = "More Details"')
        rows = self.cursor.fetchall()

        # Assert that the inserted row is present in the database
        self.assertGreater(len(rows), 0, "The newly inserted row should be found in the database.")

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
